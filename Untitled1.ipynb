Import Dependencies

```python
import tensorflow as tf
import os
import cv2
import imghdr
from matplotlib import pyplot as plt
import numpy as np
```

Clean datasets already exist for this type of classifier, but I wanted to practice scraping my own data, so I collected images of dogs and cats from Google Images using a Chrome extension that downloads all images on the page.

Now we need to clean the data by removing images that do not fit certain criteria.

First, we need to set the location of the data for access.

```python
# this is where my data is stored
data_dir = '/Users/logan/docs/DS_Final/data'
```

These are image extensions that we are wanting to use.

```python
image_exts = ['jpeg', 'jpg', 'bmp', 'png']
```

We manually remove all images from the scrape that were less than 10kb in size.

Then, to clean the data further, we go through each image and check the extension with the following method:

```python
# iterate through the directory
for image_class in os.listdir(data_dir):
    # iterate through each subdirectory
    for image in os.listdir(os.path.join(data_dir, image_class)):
        # get the path to the image
        image_path = os.path.join(data_dir, image_class, image)
        try:
            # store the image as a numpy array
            img = cv2.imread(image_path)
            # check the extension
            ext = imghdr.what(image_path)
            if ext not in image_exts:
                print('Image is not one of the permissible types {}'.format(image_path))
                os.remove(image_path)
        except Exception as e:
            print('Issue with image {}'.format(image_path))
```

Exploring the Data

First, we load in the data.

To do this, we use keras to create a data pipeline.

We build an image dataset and resizes the images:

```python
data = tf.keras.utils.image_dataset_from_directory(data_dir)
```

Then, we instantiate an iterator to allow access to the data pipeline.

```python
data_iterator = data.as_numpy_iterator()
```

We then create a "batch" variable that store the batch coming from the pipeline. 

```python
batch = data_iterator.next()
```

"batch" is an array, and images will be in "batch[0]" while labels will be in "batch[1]".

Here we explore the shape of the batch elements:

```python
shape0 = batch[0].shape
shape1 = batch[1].shape

print('Shape of batch[0]:', shape0)
print('Shape of batch[1]:', shape1)
```

Let's get an idea of how our data looks.

We can visualize a subset of our data using matplotlib.

```python
fig, axes = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    axes[idx].imshow(img.astype(int))
    # sets the title of each plot to the corresponding label
    axes[idx].title.set_text(batch[1][idx])
```

We can examine the minimum and maximum RGB value for our batch:

```python
# Min RGB value for each image
print('min: ', batch[0].min())

# Max value for each image
print('max: ', batch[0].max())
```

We want to deal with smaller input values, so we scale these "min" and "max" to be between 0 and 1.

```python
scaled_batch = batch[0] / 255
```

We can examine the "min" and "max" for our scaled batch:

```python
# Min value for each image
print('min: ', scaled_batch.min())

# Max value for each image
print('max: ', scaled_batch.max())
```

We want this transformation to occur automatically when using the pipeline.

To do this, we use a lambda function to divide our "batch[0]" values by 255 as they come in.

```python
data = data.map(lambda x, y: (x/255, y))
```

Now we use the ".as_numpy_iterator()" to convert the elements in "data" to NumPy arrays.

```python
scaled_iterator = data.as_numpy_iterator()
```

```python
batch = scaled_iterator.next()
```

Now we can visualize a subset of our data again, using our reformed "batch."

```python
fig, axes = plt.subplots(ncols=4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    axes[idx].imshow(img)
    axes[idx].title.set_text(batch[1][idx])
```

Splitting the Data Into Training and Testing

```python
print('Number of batches: ', len(data))
```

We split the batches into three groups: training, validating, and testing.

```python
train_size = int(len(data)*.8)
print('Batches for training: ', train_size)

val_size = int(len(data)*.2)
print('Batches for validation: ', val_size)

test_size = int(len(data)*.1)+1
print('Batches for testing: ', test_size)
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

Let's create more visual representations to understand our batches.

```python
labels = ['Cat', 'Dog']
counts = [sum(batch[1] == i) for i in range(2)]

plt.bar(labels, counts)
plt.title('Distribution of Cats and Dogs in This Batch')
plt.xlabel('Labels')
plt.ylabel('Counts')
plt.show()
```

```python
plt.hist(batch[0].ravel(), bins=256)
plt.title('Distribution of Pixel Intensities in This Batch')
plt.xlabel('Pixel Intensity')
plt.ylabel('Number of Pixels')
plt.show()
```

```python
data = [batch[0][batch[1]==i].ravel() for i in range(2)]

plt.boxplot(data)
plt.xticks(range(1, 3), ['Cat', 'Dog'])
plt.title('Distribution of Pixel Intensities by Label')
plt.xlabel('Labels')
plt.ylabel('Pixel Intensity')
plt.show()
```

Building the Deep Neural Network

Import dependencies

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

Here we create the model.

```python
model = Sequential([
    Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)),
    MaxPooling2D(),
    Conv2D(32, (3,3), 1, activation='relu'),
    MaxPooling2D(),
    Conv2D(16, (3,3), 1, activation='relu'),
    MaxPooling2D(),
    Flatten(),
    Dense(256, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile('adam', loss=tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

Now that we have created the model, lets use the ".summary()" method to explore and further understand our model.

```python
model.summary()
```

Train the Model

We assign a directory to store our logs for this model.

```python
logsdir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logsdir)
```

```python
history = model.fit(train, epochs=20, validation_data=val, callbacks=[tensorboard_callback])
```

Plotting Performance

Visualizing loss.

```python
figure = plt.figure()
plt.plot(history.history['loss'], color='teal', label='loss')
plt.plot(history.history['val_loss'], color='orange', label='val_loss')
figure.suptitle('Loss vs Epoch', fontsize=20)
plt.legend(loc='upper right'),
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

Visualizing accuracy.

```python
figure = plt.figure()
plt.plot(history.history['accuracy'], color='teal', label='accuracy')
plt.plot(history.history['val_accuracy'], color='orange', label='val_accuracy')
figure.suptitle('Accuracy vs Epoch', fontsize=20)
plt.legend(loc='upper left'),
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()
```

Evaluating Performance

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
precision = Precision()
recall = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X, y = batch
    yhat = model.predict(X)
    precision.update_state(y, yhat)
    recall.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print (f'Precision:{precision.result().numpy()}, Recall:{recall.result().numpy()}, Accuracy:{acc.result().numpy()}')
```
